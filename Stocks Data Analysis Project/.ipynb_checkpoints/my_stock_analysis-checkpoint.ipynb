```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path = r'C:\Users\abhis\Downloads\Compressed\Data\2-Time Series Data Analysis\individual_stocks_5yr'
company_list = ['AAPL_data.csv', 'GOOG_data.csv', 'MSFT_data.csv', 'AMZN_data.csv']

all_data = pd.DataFrame()
for file in company_list:
    current_df = pd.read_csv(path+'/'+file)
    all_data = pd.concat([all_data,current_df])

all_data.shape
```

```python
all_data.info()
```

```python
all_data['date'] = pd.to_datetime(all_data['date'])
```

```python
all_data.info()
```

```python
all_data.head()
```

```python
all_data['date'][0]
```

# Analyzing closing price of our stocks

```python
tech_list = all_data['Name'].unique()
tech_list
```

```python
i=0
plt.figure(figsize=(12,8))
for company in tech_list:
    df = all_data[all_data['Name']==company]
    i=i+1
    for date in df['date']:
        plt.subplot(2,2,i)
        plt.plot(df['date'],df['close'])
        plt.title(company)
```

```python
#Using plotly
for company in tech_list:
    df = all_data[all_data['Name']==company]
    fig = px.line(df,x='date',y='close')
    fig.show()
```

# Analyze total volume of stock being traded

```python
i=0
plt.figure(figsize=(15,8))
for company in tech_list:
    df = all_data[all_data['Name']==company]
    i+=1
    for volume in df['volume']:
        plt.subplot(2,2,i)
        plt.plot(df['date'],df['volume'])
        plt.title(company)
```

```python
#Using plotly
for company in tech_list:
    df = all_data[all_data['Name']==company]
    fig = px.line(df,x='date', y='volume',title=company)
    fig.show()
```

# Checking if stock prices are correlated

```python
tech_list
```

```python
appl_df = all_data[all_data['Name']=='AAPL']
goog_df = all_data[all_data['Name']=='GOOG']
msft_df = all_data[all_data['Name']=='MSFT']
amzn_df = all_data[all_data['Name']=='AMZN']
```

```python
close = pd.DataFrame()
close['appl'] = appl_df['close']
close['goog'] = goog_df['close']
close['msft'] = msft_df['close']
close['amzn'] = amzn_df['close']
```

```python
close.head()
```

```python
sns.pairplot(close)
```

```python
sns.heatmap(close.corr(), annot=True)
```

```python
#Closing price of Google and Microsoft are well correlated and Closing price of Amazon and Microsoft have a co-relation of 0.96(96% chance if amazon stock increase then microsoft will also increase)
```

# Daily return of each stock & how they are co-related

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
appl_df['appl_change'] = ((appl_df['close']-appl_df['open'])/appl_df['close'])*100
appl_df['goog_change'] = ((goog_df['close']-goog_df['open'])/goog_df['close'])*100
appl_df['msft_change'] = ((msft_df['close']-msft_df['open'])/msft_df['close'])*100
appl_df['amzn_change'] = ((amzn_df['close']-amzn_df['open'])/amzn_df['close'])*100
```

```python
daily_change = pd.DataFrame()
daily_change['appl_change'] = appl_df['appl_change']
daily_change['goog_change'] = appl_df['goog_change']
daily_change['msft_change'] = appl_df['msft_change']
daily_change['amzn_change'] = appl_df['amzn_change']
daily_change.head()
```

```python
sns.pairplot(daily_change)
```

```python
sns.heatmap(daily_change.corr(),annot=True)
```

```python
daily_change.corr()
```

```python
#Amazon and microsoft have good correlation on daily returns
```

# Value at Risk analysis for Apple

```python
sns.distplot(daily_change['appl_change'])
```

```python
#It follows normal distribution
```

```python
daily_change['appl_change'].std()*2
# Has 95% of entire data
```

```python
daily_change['appl_change'].std()*3
# Has 99.7% of entire data
```
