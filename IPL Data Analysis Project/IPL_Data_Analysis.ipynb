```python
def get_player(col):
    return col.split('(')[0]

def get_team(col):
    return col.split('(')[1][0:-1]
```

```python
batsman_data = batsman.copy()
```

```python
batsman_data['Team'] = batsman_data['Player'].apply(get_team)
```

```python
batsman_data['Player'] = batsman_data['Player'].apply(get_player)
```

```python
batting_order = pd.read_csv(r'D:\Untitled Folder\IPL/batting_order.csv')
```

```python
batting_order.head()
```

```python
batsman_data = pd.merge(left=batsman_data,right=batting_order,on='Player',how='left')
```

```python
batsman_data.head()
```

```python
#Replacing - with 0 in data

for i in batsman_data.columns:
    batsman_data[i] = batsman_data[i].replace(['-'],0)
```

```python
batsman_data.head()
```

```python
def r_asterisk(col):
    return col.split('*')[0]
```

```python
batsman_data.columns
```

```python
for i in batsman_data.columns[1:8]:
    batsman_data[i] = batsman_data[i].apply(r_asterisk)
```

```python
convert_dict = {
    
    'Player': object,
    'Mat' : int,
    'Inns' : int,
    'NO' : int,
    'Runs' : int,
    'HS' : int,
    'Ave' : float,
    'BF' : int,
    'SR' : float,
    'Hundred' : int,
    'Fifty' : int,
    'Duck' : int,
    'Four' : int,
    'Six' : int,
    'Team' : object
}
```

```python
batsman_data['Boundary_percent'] = ((batsman_data['Four']*4+batsman_data['Six']*6)/batsman_data['Runs'])*100
```

```python
batsman_data['Avg_bf'] = batsman_data['BF']/batsman_data['Inns']
```

# Strategy For Openers
	
|Parameters | Description | Criteria |
|:----------|:-----------:|:-------- |
|Batting Average|Avg runs scored in an innings|	>40|
|Strike Rate	|No of runs scored per 100 balls	|>140|
|Innings|	Total innings batted	|>5|
|Boundary %|	% of runs scored in boundaries	|>50|
|batting_order| |<=2|


```python
filter1 = batsman_data['Ave']>40
filter2 = batsman_data['SR']>135
filter3 = batsman_data['Inns'] > 7
filter4 = batsman_data['Boundary_percent'] > 50
filter5 = batsman_data['BO']<=2
```

```python
openers = batsman_data.where(filter1 & filter2 & filter3 & filter4 & filter5)
```

```python
openers = openers.dropna()
```

```python
openers.head(3)
```

# Strategy For Middle Order
	
|Parameters | Description | Criteria |
|:----------|:-----------:|:-------- |
|Batting Average|Avg runs scored in an innings|	>40|
|Strike Rate	|No of runs scored per 100 balls	|>120|
|Innings|	Total innings batted	|>5|
|Avg_bf|	Avg balls faced in an innings	|>20|
|batting_order| |>2|


```python
filter1 = batsman_data['Ave']>35
filter2 = batsman_data['SR']>120
filter3 = batsman_data['Inns'] > 5
filter4 = batsman_data['Avg_bf']>19
filter5 = batsman_data['BO'] > 2
filter6 = batsman_data['BO'] <=5
```

```python
middle_order = batsman_data.where(filter1 & filter2 & filter3 & filter4 & filter5 & filter6)
```

```python
middle_order = middle_order.dropna()
middle_order.sort_values(by = 'BO')
```

# Strategy For Finisher/Lower Order
	
|Parameters | Description | Criteria |
|:----------|:-----------:|:-------- |
|Batting Average|Avg runs scored in an innings|	>25|
|Strike Rate	|No of runs scored per 100 balls	|>150|
|Innings|	Total innings batted	|>3|
|Avg_bf|	Avg balls faced in an innings	|>8|
|batting_order| |>5|


```python
filter1 = batsman_data['Ave']>25
filter2 = batsman_data['SR']>150
filter3 = batsman_data['Inns'] > 3
filter4 = batsman_data['Avg_bf']>=8
filter5 = batsman_data['BO']>5
```

```python
finisher = batsman_data.where(filter1 & filter2 & filter3 & filter4 & filter5)
```

```python
finisher.dropna(inplace=True)
```

```python
finisher.head()
```

```python
openers.to_csv('openers.csv',index=False)
middle_order.to_csv('middle_order.csv',index=False)
finisher.to_csv('finisher.csv',index=False)
```
